```python
# Connect to Website and pull in data
# Find Your User-Agent: https://httpbin.org/get

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# headless background execution
PATH = "C:\Program Files (x86)\chromedriver.exe"
Options = Options()
Options.headless = True




# url = col_list[25]
url = "https://articulo.mercadolibre.com.mx/MLM-1444756143-una-vida-sin-salud-mental-alba-gonzalez-_JM#position=6&search_layout=grid&type=item&tracking_id=31741b42-f099-4bff-8703-af545bc46bdb&DEAL_ID=MLM12933&S=landingHublibros-revistas-y-comics&V=13&T=CarouselDynamic-home&L=LOS-MAS-VENDIDOS&deal_print_id=82805ce0-33e2-11ee-8a43-5baddc9422a8&c_id=carouseldynamic-home&c_element_order=undefined&c_campaign=LOS-MAS-VENDIDOS&c_uid=82805ce0-33e2-11ee-8a43-5baddc9422a8"

headers = {
    'User-Agent': ('Mozilla/5.0 (X11; Linux x86_64)'
                    'AppleWebKit/537.36 (KHTML, like Gecko)'
                    'Chrome/44.0.2403.157 Safari/537.36'),
    'Accept-Language': 'en-US, en;q=0.5'
}

# url = "https://pokedex.org/"
browser = webdriver.Chrome(options=Options)
browser.get(url)
print(browser.title)
# browser.quit()
page = browser.page_source
# page = BeautifulSoup(requests.get(url).content, 'lxml')
# uls = html.find('ul', attrs={'id': 'monsters-list'})
# print(uls.prettify())
# page = requests.get(url, headers=headers)

soup1 = BeautifulSoup(page, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

# print(page)
print(soup2)
```

```python
title = soup2.find_all("h1", {"class": "ui-pdp-title"})[0].get_text().strip()
price = soup2.find_all("span", {"class": "andes-money-amount__fraction"})[0].string.strip()
rating = soup2.find_all("span", {"class": "ui-pdp-review__amount"})
description = soup2.find_all("p", {"class": "ui-pdp-description__content"})[0].get_text()
features = soup2.find_all("li", {"class": "ui-pdp-features__item"})

# Create a Timestamp for your output to track when data was collected
today = datetime.date.today()

print(title)
print(price)
print(rating)
print(description)
print(features)
print(today)
```

```python
# Create CSV and write headers and data into the file

import csv 

header = ['Title', 'Price', 'Date']
data = [title, price, today]


with open('AmazonWebScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
    
```
